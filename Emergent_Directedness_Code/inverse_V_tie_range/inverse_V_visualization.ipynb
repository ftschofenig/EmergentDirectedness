```python
import matplotlib.pyplot as plt
import numpy as np

# List of dataframes and labels
dataframes = [df_clustered_power_law, df_AddHealth, df_Banerjee]
labels = ['Clustered Power Law Network', 'Add Health Dataset', 'Banerjee Village Networks']
colors = ['#d62728', '#ff7f0e', '#2ca02c']  # Light blue, orange, and green for better contrast

plt.figure(figsize=(4, 7))

# Initialize offset for each dataset
bar_width = 0.1
num_groups = len(dataframes)

# Loop through each dataframe and plot its mean, std, and connecting line
all_x_positions = []
all_means = []
for i, (df, label, color) in enumerate(zip(dataframes, labels, colors)):
    grouped = df.groupby('neighborhood_overlap_relative_quantile')['CPC_sum']
    means = grouped.mean()
    stds = grouped.std()
    
    # Convert index to numeric positions with slight shifts
    x_positions = np.arange(len(means)) + (i - 1) * bar_width  # Shift each dataset slightly
    all_x_positions.append(x_positions)
    all_means.append(means)
    
    # Plot error bars (thin lines, no caps, light color)
    plt.errorbar(x_positions, means, yerr=stds, fmt='o', color=color, 
                 capsize=0, elinewidth=1, markeredgewidth=1, label=label)
    
    # Draw dotted lines connecting the same dataset points across tie types
    plt.plot(x_positions, means, linestyle='dotted', color=color, alpha=0.7)

# Remove top and right axis lines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Customize plot
plt.xlabel('Tie type', fontsize=14)
plt.ylabel('Mean Tie Importance', fontsize=14)
plt.xticks(ticks=np.arange(len(means)), labels=means.index, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12, frameon=False, loc='upper center', bbox_to_anchor=(0.5, 1.5))
plt.grid(False)  # No grid for minimalist style

plt.tight_layout()
plt.savefig("Average_Tie_Importance_by_Tie_Strength_Group.png", dpi=300)
plt.show()
```

```python
#accumulate the degree distributions of the graphs in Banerjee_graphs, AddHealth_Graphs and graphs and plot them separately in a histogram
degree_distributions_Banerjee, degree_distributions_AddHealth, degree_distributions_Powerlaw = {}, {}, {}
for name, graph in Banerjee_graphs.items():
    degrees = [d for n, d in graph.degree()]
    degree_distributions_Banerjee[name] = degrees
for name, graph in AddHealth_Graphs.items():
    degrees = [d for n, d in graph.degree()]
    degree_distributions_AddHealth[name] = degrees
for i, graph in enumerate(graphs):
    degrees = [d for n, d in graph.degree()]
    degree_distributions_Powerlaw[f"Powerlaw_{i}"] = degrees
#plot the degree distributions in a histogram
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
all_degrees = [d for degrees in degree_distributions_Banerjee.values() for d in degrees]
plt.hist(all_degrees, bins=range(1, max(all_degrees)+1), color='green', alpha=0.3)
all_degrees = [d for degrees in degree_distributions_AddHealth.values() for d in degrees]
plt.hist(all_degrees, bins=range(1, max(all_degrees)+1), color='red', alpha=0.3)
all_degrees = [d for degrees in degree_distributions_Powerlaw.values() for d in degrees]
#plt.hist(all_degrees, bins=range(1, max(all_degrees)+1), color='blue', alpha=0.3)
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.xlim(0, 50)
plt.legend(["Banerjee", "AddHealth"])
#plt.yscale('log')
#plt.xscale('log')
plt.grid(False)
```
